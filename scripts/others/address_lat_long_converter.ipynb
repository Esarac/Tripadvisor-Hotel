Import

```python
import pandas as pd
from geopy.extra.rate_limiter import RateLimiter
from geopy.geocoders import Nominatim
```

Load csv

```python
DATA_PATH = '../../data/qualitative_analysis/sentiment/tripadvisor_hotels_summaries_whr_sentiments-small.csv'
df = pd.read_csv(DATA_PATH)
df
```

Lat and Long method

```python
locator = Nominatim(user_agent="myGeocoder")

geocode = RateLimiter(locator.geocode, min_delay_seconds=1)

df_location = df.copy()
df_location['HOTEL_ADDRESS'] = df_location['HOTEL_LOCATION']

print("start")
df_location['HOTEL_LOCATION'] = df_location['HOTEL_LOCATION'].apply(geocode)
print("end")

df_location['HOTEL_POINT'] = df_location['HOTEL_LOCATION'].apply(lambda loc: tuple(loc.point) if loc else None)

df_location[['HOTEL_LATITUDE', 'HOTEL_LONGITUDE', 'HOTEL_ALTITUDE']] = pd.DataFrame(df_location['HOTEL_POINT'].tolist(), index=df.index)

df_location
```

```python
OUTPUT_PATH = '../../data/address_converter/tripadvisor_hotels_data-raw.csv'
df_location.to_csv(OUTPUT_PATH, index=False)
```

```python
df_nolatlon = df_location.loc[df_location['HOTEL_LONGITUDE'].isnull()].groupby(["HOTEL_NAME", "HOTEL_ADDRESS"]).size().reset_index(name='HOTEL_REVIEW_COUNT') # COUNT
df_nolatlon['HOTEL_ADDRESS']
df_nolatlon
```

```python
OUTPUT_NO_LATLONG_PATH = '../../data/address_converter/tripadvisor_hotels_no_lat_lon.csv'
df_nolatlon.to_csv(OUTPUT_NO_LATLONG_PATH, index=False)
```

## Easy Way

Import

```python
import pandas as pd
```

Load csv

```python
DATA_PATH = '../../data/qualitative_analysis/sentiment/tripadvisor_hotels_summaries_whr_sentiments-small.csv'
df = pd.read_csv(DATA_PATH)
df
```

Get all hotel addresses

```python
df_hotels = df.groupby(["HOTEL_NAME", "HOTEL_LOCATION"]).size().reset_index(name='HOTEL_REVIEW_COUNT') # COUNT
df_hotels
```

Save

```python
OUTPUT_HOTELS_PATH = '../../data/address_converter/output/hotel_addresses.csv'
df_hotels.to_csv(OUTPUT_HOTELS_PATH, index=False)
```

Then, we upload the generated csv file into the [geoapify website](https://www.geoapify.com/tools/geocoding-online).

After that, we import the new csv file.

```python
HOTEL_COORDS_PATH = '../../data/address_converter/input/hotel_coords.csv'
df_coords = pd.read_csv(HOTEL_COORDS_PATH)
```

Join the two df

```python
df_coords = df_coords[['original_HOTEL_NAME','lat', 'lon','formatted']]


df_new = pd.merge(df, df_coords, how='inner', left_on = 'HOTEL_NAME', right_on = 'original_HOTEL_NAME')
```

Adjust df

```python
df_new = df_new.drop(columns=['original_HOTEL_NAME', 'HOTEL_LOCATION']).rename(columns={"lat": "LATITUDE", "lon": "LONGITUDE", "formatted":"ADDRESS"})
```

Save csv

```python
OUTPUT_PATH = '../../data/address_converter/output/tripadvisor_hotels_data-clean.csv'
df_new.to_csv(OUTPUT_PATH, index=False)
```
